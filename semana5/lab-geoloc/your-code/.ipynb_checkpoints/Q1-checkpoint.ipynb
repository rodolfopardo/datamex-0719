El primer paso es seleccionar lo datos desde MongoDB, pasarlos a un dataframe, limpiarlos y guardar las seleccion de datos en un archivo json para volver a llevarlos a MongoDB.

```python
# Tu codigo aqui

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
!{sys.executable} -m pip install pymongo
from pandas.io.json import json_normalize
```

```python
#Conexion a la base de Mongo

from pymongo import MongoClient
client=MongoClient()
db=client.Ironhack
```

```python
#Vamos a extraer las primeras bases 

m = db.Ironhack.find({'number_of_employees': {'$lt': 50}},\
                      {'name':1, 'number_of_employees': 1, '_id':0, 'offices':1})\
                      .sort('number_of_employees', -1)
```

```python
#Lo convertimos a lista de jsons
m = list(m)
m
```

```python
#Normalizamos nuestra lista de json's

offices = json_normalize(m, 'offices')
```

```python


offices = offices.dropna(subset=['latitude', 'longitude'])
```

```python
offices.isnull().sum()
```

```python
offices.shape
```

```python
#Primeros países con mayor cantidad de startups  

offices.groupby('country_code')['zip_code'].count().sort_values(ascending = False).head(4)
```

Estados Unidos, Gran Bretaña, Canadá y Alemania son los países con mayor cantidad de startups en todo el mundo

```python
#Primeras ciudades con mayor cantidad de startups 

offices.groupby('city')['zip_code'].count().sort_values(ascending = False).head(3)
```

San Francisco, New York y London son las ciudades con mayor cantidad de startups

```python
#offices['location'] = offices[['longitude', 'latitude']].values.tolist()
offices['Geolocation'] = offices.apply(lambda x: [x.longitude, x.latitude], axis=1)
```

```python
offices
```

```python
#DataFrame to json 

offices.to_json('startups2.json',orient='records')
offices.to_csv('startups2.csv')
```
